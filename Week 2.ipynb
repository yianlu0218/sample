```python
np.random.seed(1000) #
y = np.random.standard_normal(20)
x = np.arange(len(y))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 6))#圖的尺寸
plt.plot(y.cumsum(), 'k', lw=1.5) #b代表藍色
plt.plot(y.cumsum(), 'bo') #ro好像是代表紅點
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```
